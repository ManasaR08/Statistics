## Checking if a distribution is a Gaussian Distribution Using Q-Q Plot

### Testing for a normal distribution

```python
# Checking if a distribution is Gaussian Distributed
# QQ plot

import numpy as np
import statsmodels.api as sm
import pylab as py

data_points = np.random.normal(0,1,100)

sm.qqplot(data_points, line = '45')
py.show()
```

#### If majority of the data points lie on the straight line, they are said to have a gaussian distribution.

```python
import seaborn as sns

sns.displot(data_points, kind = "kde")
```

### Testing for a Log Normal Distribition

```python
log_normal = np.random.lognormal(0,1,100)
```

```python
import seaborn as sns

sns.displot(log_normal, kind = "kde")
```

```python
sm.qqplot(log_normal, line = '45')
py.show()
```

#### Majority of the data points do not lie on the straight line. Therefore it is clear that it does not have a gaussian distribution

### Converting a Log Normal distribution to a Gaussian distribution

```python
converted = np.log(log_normal)
```

```python
sm.qqplot(converted, line = '45')
py.show()
```

```python
sns.displot(converted, kind = "kde")
```

#### The log normal distribution has been converted into gaussian distribution by applying log transformation.
