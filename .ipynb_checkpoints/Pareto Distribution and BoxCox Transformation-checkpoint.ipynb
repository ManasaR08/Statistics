```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#create Pareto Distribution
data = np.random.exponential(size = 1000)
```

```python
#data
```

```python
#converting Pareto into Gaussian distribution using BoxCox transformation
fitted_data, fitted_lambda =  stats.boxcox(data)
```

```python
fig,ax = plt.subplots(1,2)

sns.distplot(data, hist = False, kde = True, label = "Pareto", ax = ax[0])
sns.distplot(fitted_data, hist = False, kde = True, label = "Gaussian", ax = ax[1])

fig.set_figheight(5)
fig.set_figwidth(10)

plt.legend(loc = "upper right")
```
